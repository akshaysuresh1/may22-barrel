### Model Comparison using the SMAPE metric

Let's visualize the SMAPE values returned by our models on the test data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
# Read results of simple linear regression.
df_slr = pd.read_csv('smape_slr.csv')

# Read results of random forest regression.
df_rfr = pd.read_csv('smape_rfr.csv')

# Read results of support vector regression.
df_svr = pd.read_csv('smape_svr.csv')
```

```python
df_rfr
```

```python
# Number of clusters
N_clusters = len(df_slr)
```

```python
# Plotting
#plt.rcParams['axes.grid'] = True

width = 0.2 # Width of each bar

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18,6))
# Bars for SLR
l1 = axes[0].bar(df_slr['Cluster']-width, df_slr['N_SMAPE_slr'], width=width, color='lightcoral')
axes[1].bar(df_slr['Cluster']-width, df_slr['P_SMAPE_slr'], width=width, color='lightcoral')
axes[2].bar(df_slr['Cluster']-width, df_slr['K_SMAPE_slr'], width=width, color='lightcoral')

# Bars for RFR
l2 = axes[0].bar(df_rfr['Cluster'], df_rfr['N_SMAPE_rfr'], width=width, color='lightblue')
axes[1].bar(df_rfr['Cluster'], df_rfr['P_SMAPE_rfr'], width=width, color='lightblue')
axes[2].bar(df_rfr['Cluster'], df_rfr['K_SMAPE_rfr'], width=width, color='lightblue')

# Bars for SVR
l3 = axes[0].bar(df_svr['Cluster']+width, df_svr['N_SMAPE_svr'], width=width, color='darkslateblue')
axes[1].bar(df_svr['Cluster']+width, df_svr['P_SMAPE_svr'], width=width, color='darkslateblue')
axes[2].bar(df_svr['Cluster']+width, df_svr['K_SMAPE_svr'], width=width, color='darkslateblue')

# Subplot titles
axes[0].set_title('N consumption (kg/ha)', fontsize=16)
axes[1].set_title('P consumption (kg/ha)', fontsize=16)
axes[2].set_title('K consumption (kg/ha)', fontsize=16)
# x-axis labels
axes[0].set_xlabel('Cluster number', fontsize=14)
axes[1].set_xlabel('Cluster number', fontsize=14)
axes[2].set_xlabel('Cluster number', fontsize=14)
# y-axis labels
axes[0].set_ylabel('SMAPE (%) on test data', fontsize=14)
axes[1].set_ylabel('SMAPE (%) on test data', fontsize=14)
axes[2].set_ylabel('SMAPE (%) on test data', fontsize=14)
# x-axis ticks
axes[0].set_xticks(np.arange(1,N_clusters+1))
axes[1].set_xticks(np.arange(1,N_clusters+1))
axes[2].set_xticks(np.arange(1,N_clusters+1))
# tick sizes
axes[0].tick_params(axis='x', labelsize=14)
axes[0].tick_params(axis='y', labelsize=14)
axes[1].tick_params(axis='x', labelsize=14)
axes[1].tick_params(axis='y', labelsize=14)
axes[2].tick_params(axis='x', labelsize=14)
axes[2].tick_params(axis='y', labelsize=14)
# y-axis limits
axes[0].set_ylim((0,100))
axes[1].set_ylim((0,100))
axes[2].set_ylim((0,100))

# Figure legend
fig.legend((l1,l2,l3), ('Simple linear regression', 'Random forest regression', 'Support vector regression'), 
           loc=(0.25,0.94), fontsize=14,  ncol=3)
fig.subplots_adjust(left=0.01, right=0.98, wspace=0.2)
plt.show()
```
