# We creat Heatmap correlation matrices with respect to clustered data


## _KMeans_ heatmap correlation matrices

We creat correlation matrix heatplots  with respect to the parameters described below for each of the KMeans clusters for k=5

The parameters:

`prec_grow` <br>
 `prec_harvest`<br>
 `et_grow`<br>
 `et_harvest`<br>
 `maxT_grow`<br>
 `maxT_harvest`<br>
 `minT_grow`<br>
 `minT_harvest`<br>
 `windspeed_grow`<br>
 `windspeed_harvest`<br> 
 `runoff_grow`<br>
 `runoff_harvest`<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
import seaborn
from seaborn import set_style
from pandas.plotting import scatter_matrix
import os

set_style("whitegrid")

from sklearn.cluster import KMeans
```

```python
kmean_cluster = pd.read_csv('../Final_data/rice_yield_KMeans_averaged.csv')
```

```python
k = 6
corr_matrices = []
for i in range(k):
    df = kmean_cluster[kmean_cluster["KM5_clusterN_(maxT_growth_maxT_harvest)"] == i]
    
    corr_m = df.corr()
    corr_matrices.append(corr_m)
   
```

```python
fig = plt.figure( figsize = (16, 50))

for i in range(1, k+1):
    plt.subplot(k, 1, i)
    seaborn.heatmap(corr_matrices[i-1] , annot=True)
    
plt.show()
```

## Here we create similar heatmap correlation matrices but with respect to the _hierarchical clustering_

```python

params = ["prec_grow",
                    "prec_harvest",
                    "et_grow",
                    "et_harvest",
                    "maxT_grow",
                    "maxT_harvest",
                    "minT_grow",
                    "minT_harvest",
                    "windspeed_grow",
                    "windspeed_harvest",

                    'runoff_grow', 
                    'runoff_harvest',
                    'Cluster']
```

```python
hierarch_cluster = pd.read_csv('../Final_data/rice_yield_clusters.csv')

hierarch_cluster = hierarch_cluster[params]
```

```python
n = 6
corr_matrices_h = []
for i in range(1, n+1):
    df = hierarch_cluster[hierarch_cluster["Cluster"] == i]
    
    corr_m_h = df.corr()
    corr_matrices_h.append(corr_m_h)
   
```

```python
fig = plt.figure( figsize = (16, 60))

for i in range(1, n):
    plt.subplot(n, 1, i)
    seaborn.heatmap(corr_matrices_h[i] , annot=True)
    
plt.show()
```
